# Importing Libraries

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# 1.Loading Data

```python
data_2021_2022 = pd.read_json("my_spotify_data/spotify_extended_streaming_history/Streaming_History_Audio_2021-2022_0.json")
data_2021_2022.head()
```

```python
data_2022_2023 = pd.read_json("my_spotify_data/spotify_extended_streaming_history/Streaming_History_Audio_2022-2023_1.json")
data_2022_2023.head()
```

```python
data_2023_2024 = pd.read_json("my_spotify_data/spotify_extended_streaming_history/Streaming_History_Audio_2023-2024_2.json")
data_2023_2024.head()
```

# 2.Start of Cleaning Process

**Merging Datasets**

```python
merged_dataset = pd.concat([data_2021_2022, data_2022_2023, data_2023_2024], ignore_index=True)
merged_dataset.head()
```

**Checking For Missing Values**

```python
missing_values = merged_dataset.isnull().sum()
print("Missing Values:\n", missing_values)
```

**Dropping Columns With Missing Values and Sensitive Info**

```python
# Dropping specific columns from merged_dataset
clean_data = merged_dataset.drop(["episode_name", "episode_show_name", "spotify_episode_uri", "user_agent_decrypted", "skipped", "ip_addr_decrypted", "username"], axis=1)

# Displaying the modified DataFrame
clean_data.head()
```

**Exporting Cleaned Data to a json File**

```python
save_clean_data = clean_data.to_json("my_spotify_data/spotify_extended_streaming_history/clean_data.json", index=False)
```

**Convert Timestamp Columns to Datetime Objects**

```python
# Convert timestamp columns to datetime objects
clean_data['ts'] = pd.to_datetime(clean_data['ts'])

# Ensure consistency in the timestamp format
print("Timestamps after conversion:\n", clean_data['ts'].head())
```

**Identify and handle anomalies or outliers in the data**

```python
# Identify anomalies (for example, consider ms_played values less than 0 as anomalies)
anomalies = clean_data[clean_data['ms_played'] < 0]
print("Anomalies:\n", anomalies)

# Remove anomalies
clean_data = clean_data[clean_data['ms_played'] >= 0]

```

```python
clean_data.head()
```

# 3.Data Exploration

**Distribution Analysis: Streaming Durations**

```python
plt.figure(figsize=(12, 6))
sns.histplot(clean_data['ms_played'], bins=50, kde=True)
plt.title('Distribution of Streaming Durations')
plt.xlabel('Milliseconds Played')
plt.ylabel('Frequency')
plt.show()
```

**Distribution Analysis: Platforms**

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='platform', data=clean_data)
plt.title('Distribution of Streaming Platforms')
plt.xlabel('Platform')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

**Distribution Analysis: Countries**

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='conn_country', data=clean_data)
plt.title('Distribution of Streaming Countries')
plt.xlabel('Country')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

**Identifying: Top 10 Tracks**

```python
top_tracks = clean_data['master_metadata_track_name'].value_counts().head(10)
print("Top 10 Tracks:\n", top_tracks)
```

**Identifying: Top 10 Artists**

```python
top_artists = clean_data['master_metadata_album_artist_name'].value_counts().head(10)
print("\nTop 10 Artists:\n", top_artists)
```

**Identifying: Top 10 Albums**

```python
top_albums = clean_data['master_metadata_album_album_name'].value_counts().head(10)
print("\nTop 10 Albums:\n", top_albums)
```

**Analyzing Streaming Habits Over Time**

```python
# Analyze streaming habits over time (monthly trends)
monthly_streaming_trends = clean_data.resample('M', on='ts')['ms_played'].sum()

# Plot streaming trends
plt.figure(figsize=(14, 6))
monthly_streaming_trends.plot(marker='o')
plt.title('Monthly Streaming Trends Over Time')
plt.xlabel('Date')
plt.ylabel('Milliseconds Played')
plt.show()
```
